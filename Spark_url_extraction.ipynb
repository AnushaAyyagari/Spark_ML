```python
from pyspark.sql import SparkSession

spark = SparkSession.builder\
        .master("local")\
        .appName("webscrapping")\
        .config('spark.ui.port', '4050')\
        .getOrCreate()
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
'''
i am using python libraries like requests and beautiful Soup for webscarpping
'''

url='https://www.nba.com/'



headers = {'User-Agent':
        'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.93 Safari/537.36'}

page = requests.get(url, headers = headers, timeout = 5)

print(page.status_code)

soup = BeautifulSoup(page.content, 'html.parser')

#soup.body()
```

```python
tab=soup.findAll("img",{"alt":True, "src":True})
tab
```

```python
'''
creating a list of all the image url's
'''
url_list=[]
for image in tab:
      url_list.append(image['src'])
```

```python
'''
The only refence for this part of the code was from the blackboard slides
in thhis step i am converting the processed data into an RDD
'''
url_RDD=spark.sparkContext.parallelize(url_list)
```

```python
'''
to confirm the object type
'''
print(type(url_RDD))
```

```python
lst_count=url_RDD.count()
print('there are {} images in {} site'.format(lst_count,url))
```

```python
data_display=url_RDD.collect()
print('the images are:')
for links in data_display:
  print(links)
```

```python
path='/content/sample_data/california_housing_test.csv'
csv_RDD=spark.read.csv(path).rdd
print(type(csv_RDD))
data=csv_RDD.collect()
for my_data in data:
  print(my_data)
```

```python
path='/content/sample_data/anscombe.json'
json_RDD=spark.read.json(path).rdd
print(type(json_RDD))
jason_data=json_RDD.collect()
for j_data in jason_data:
  print(j_data)
filte1_rdd = json_RDD.map(lambda x: (x[1],x[0]))

filter2=json_RDD.flatMap(lambda x: [(item,x[0]) for item in x[1]])
print(filter2)

reduced_value=json_RDD.reduce(lambda x,y: x+y)
print(reduced_value)
```
